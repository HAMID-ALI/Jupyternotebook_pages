```python
import pandas
```

```python
import geopy
```

```python
df1 = pandas.read_json("http://pythonhow.com/supermarkets.json")
```

```python
df2 = df1.set_index("ID")
```

```python
df2
```

```python
df2["Address"] = df2["Address"] + ", " + df2["City"] + ", " + df2["State"]
```

```python
df2
```

```python
from geopy.geocoders import Nominatim
```

```python
nom = Nominatim()
```

```python
df2["Coordinates"] = df2["Address"].apply(nom.geocode)
```

```python
df2
```

```python
df2["Latitude"] = df2["Coordinates"].apply(lambda x: x.latitude if x!=None else None)
```
